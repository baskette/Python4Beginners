```python
# let's plot a simple line graph
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
plt.plot(input_values, squares, linewidth=5)

# set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
plt.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
# plottiing a single point using simple scatter

plt.scatter(2, 4, s=200)

# set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# Plotting a seies of points with scatter()
# we start by passing scatter() seperate x- and y- values

x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.scatter(x_values, y_values, s=100)

# set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# what you want to calculate data automatically without making long lists

x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

# to remove outlines from the output use the argument edgecolor='none'
# to change color, pass the argument'c' to scatter() with the name of the color you want to use. Default color is blue
# you can use colors from the RGB color model. e.g c=(0,0,0.8) for a light blue. 
# values closer to 0 produce dark colors, and those closer to 1 produce lighter colors

plt.scatter(x_values, y_values, c='red', edgecolor='none', s=40)

# set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

# Set the range for each axis.
plt.axis([0, 1100, 0, 1100000])

plt.savefig('squares_plot.png', bbox_inches='tight')
```

```python
x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

# The pyplot module includes built in colormaps.
# Colormaps are series of colors in a gradient that move from a starting color to an ending color.
# Lets assign each point a color based on the value of its y-value

plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Greens, edgecolor='none', s=40)

# set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis='both', which='major', labelsize=14)

# Set the range for each axis.
plt.axis([0, 1100, 0, 1100000])

plt.show()
```
