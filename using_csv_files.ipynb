```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime

filename = 'sitka_weather_2018_full.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)


    # extract dates and temperature highs from file
    
    dates, tmaxes, tmins = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[2], "%Y-%m-%d")
            tmax = int(row[8])
            tmin = int(row[9])
        except ValueError:
            print(current_date, 'missing data')
        else:
            dates.append(current_date)
            tmaxes.append(tmax)
            tmins.append(tmin)
        

# now let's create visualization for this chart

# Plot data.
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, tmaxes, c='red')

# Format plot.
plt.title("Daily high temperatures - 2018\nSitka, Alaska", fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)

plt.show()
```

```python
# Plot data.
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, tmaxes, c='red', alpha=0.5)
plt.plot(dates, tmins, c='blue', alpha=0.5)
plt.fill_between(dates, tmaxes, tmins, facecolor='blue', alpha=0.1)

# Format plot.
plt.title("Daily high temperatures - 2018\nSitka, Alaska", fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)

plt.show()
```
