```python
# regression charts with seaborn

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style("whitegrid")
sns.regplot(x="total_bill", y="tip", data=tips)
sns.despine(left=True,bottom=True)
```

```python
# there is a positive linear relationship between the total bill and tips
# to segregate this by sex, you need to add specify 'col'=sex but use lmplot instead of regplot

sns.lmplot(x="total_bill", y="tip", data=tips, col="sex", row="time")
```

```python
# converting y axis to log scale
name = ['Steven', 'Bob', 'Juan', 'Michael', 'Bobby', 'Michelle', 'Peter', 'Blake', 'Paul', 'Mary']
salary = [10000000, 8000000, 7500000, 500000, 250000, 100000, 70000, 50000, 20000, 5000]
```

```python
sns.barplot(name,salary)
```

```python
# as you can see blake, paul and mary don't see to have a salary. we can solve this by converting y axis to log scale

g = sns.barplot(name, salary)
g.set_yscale("log")
```
